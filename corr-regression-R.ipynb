```r
dataframe <- read_csv("OA 6.6 - dating.csv")
## A
ggscatter(dataframe, x = "Icecream", y = "Games",add = "reg.line", conf.int = TRUE,cor.coef = TRUE, cor.method = "pearson",xlab = "Icecream", ylab = "Games")
## P = 0.78, there is a correlation between both icecream and games
ggscatter(dataframe, x = "Games", y = "Miles",add = "reg.line", conf.int = TRUE,cor.coef = TRUE, cor.method = "pearson",xlab = "Games", ylab = "Miles")
## P = 2.2e^-16 is very small - no correlation
```

```r
# Taking two column
# Vectors with same length
x = dataframe$Icecream
y = dataframe$Games
 
# Calculating
# Correlation coefficient
# Using cor() method
result = cor(x, y, method = "pearson")
 
# Print the result
cat("Person correlation coefficient is:", result)
 
# Using cor.test() method
res = cor.test(x, y, method = "pearson")
print(res)
```

```r
## B
#~ defines relationship between 2 variables
# Apply the lm() function.
relation <- lm(formula = Games ~ Miles, data = dataframe)
print(relation)

ggplot(dataframe, aes(x=Miles,y=Games)) + geom_point()+  stat_smooth(method="lm") + ylim(0,10)
# line equation: games = 3.532e+00 + 9.003e-05*miles
```
